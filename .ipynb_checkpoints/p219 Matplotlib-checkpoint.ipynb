```python
%matplotlib inline
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.style.use('classic')

# Export and save plot
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
fig.savefig('my_figure.png')

# Import Image
from IPython.display import Image
Image('my_figure.png')

# fig.canvas.get_supported_filetypes()
```

```python
plt.figure()

plt.subplot(2, 1, 1) #rows, cols, panel_no
plt.plot(x, np.sin(x))
plt.subplot(2, 1, 2) #rows, cols, panel_no
plt.plot(x, np.cos(x))
```
