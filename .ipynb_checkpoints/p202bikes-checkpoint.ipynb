```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mp
import seaborn; seaborn.set()
import matplotlib.pyplot as plt

df = pd.read_csv('FremontBridge.csv', index_col = 'Date', parse_dates = True)

# print(df.head())

df.columns = ['West', 'East']
#df['Total'] = df.West + df.East
df['Total'] = df.eval('West + East')

# print(df.describe())
# print(df.dropna().describe())

# df.plot()
# plt.ylabel('Hourly bike count')

weekly = df.resample('W').sum()
weekly.plot(style = [':', '-', '--'])
plt.ylabel('weekly bikes');

daily = df.resample('D').sum()
daily.rolling(30, center = True).sum().plot(style = [':', '-', '--'])

daily.rolling(50, center=True, win_type='gaussian').sum(std=10).plot(style=[':', '--', '-']);

by_time = df.groupby(df.index.time).mean()
hourly_ticks = 4 * 60 * 60 * np.arange(6)
by_time.plot(xticks=hourly_ticks, style=[':', '--', '-'])

by_weekday = df.groupby(df.index.dayofweek).mean()
by_weekday.index = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
by_weekday.plot()
```
